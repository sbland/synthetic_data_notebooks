```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from dataclasses import dataclass
from enum import Enum
```

```python
class Gender(Enum):
    FEMALE = 0
    MALE = 1
    OTHER = 2


@dataclass
class Person:
    id: int
    age: int
    height: int
    gender: Gender
    unitsOfAlcohol: int
    lifeExpectancy: int
```

```python
people = []

for i in range(100):
    person = Person(id=i+1, age=np.random.randint(18, 80), height=np.random.randint(150, 200),
                    gender=np.random.choice(list(Gender)), unitsOfAlcohol=np.random.choice([0, 1, 2, 3, 4, 5]),
                    lifeExpectancy=np.random.randint(60, 100))
    people.append(person)

```

```python
ages = [person.age for person in people]
heights = [person.height for person in people]

plt.scatter(ages, heights)
plt.xlabel('Age')
plt.ylabel('Height')
plt.title('Relationship between Height and Age')
plt.show()
```

```python
units_of_alcohol = [person.unitsOfAlcohol for person in people]
life_expectancy = [person.lifeExpectancy for person in people]

plt.scatter(units_of_alcohol, life_expectancy)
plt.xlabel('Units of Alcohol')
plt.ylabel('Life Expectancy')
plt.title('Relationship between Units of Alcohol and Life Expectancy')
plt.show()
# We would expect there to be a relationship between the amount of alcohol consumed and life expectancy. The more alcohol consumed, the lower the life expectancy.
# This is not currently captured by the synthetic data we have generated. We would need to adjust the life expectancy based on the amount of alcohol consumed.
```

```python
# average_life_expectancy = np.mean([person.lifeExpectancy for person in people])
for person in people:
    person.lifeExpectancy -= person.unitsOfAlcohol * 4
```

```python
units_of_alcohol = [person.unitsOfAlcohol for person in people]
life_expectancy = [person.lifeExpectancy for person in people]

plt.scatter(units_of_alcohol, life_expectancy)
plt.xlabel('Units of Alcohol')
plt.ylabel('Life Expectancy')
plt.title('Relationship between Units of Alcohol and Life Expectancy')
plt.show()
# We would expect there to be a relationship between the amount of alcohol consumed and life expectancy. The more alcohol consumed, the lower the life expectancy.
# This is now captured by the synthetic data we have generated.
```
